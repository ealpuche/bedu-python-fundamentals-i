```python
promedio_estres = df['nivel_estres'].mean()
maximo_estres = df['nivel_estres'].max()
minimo_estres = df['nivel_estres'].min()
media_estres = df['nivel_estres'].median()

print(f'Promedio de estres: {promedio_estres}')
print(f'Maximo de estres: {maximo_estres}')
print(f'Minimo de estres: {minimo_estres}')
print(f'Media de estres: {media_estres}')
```

```python
descripcion_estres = ['Alto' if nivel_estres > media_estres else 'Bajo' for nivel_estres in df['nivel_estres']]
df['descripcion_estres'] = descripcion_estres
df
```
