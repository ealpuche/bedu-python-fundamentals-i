```python
import pandas as pd
import numpy as np

# Crear un DataFrame para un restaurante con valores faltantes
data = {
    'Orden': [101, 102, 103, 104, 105, 106, 107, 108],
    'Platillo': ['Tacos', 'Burrito', np.nan, 'Enchiladas', 'Tacos', 'Quesadilla', 'Tacos', 'Burrito'],
    'Precio': [10, 12, 11, 13, 10, 9, np.nan, 12],
    'Categoría': ['Comida rápida', 'Comida rápida', 'Comida rápida', 'Comida tradicional', 'Comida rápida', 'Comida rápida', 'Comida rápida', 'Comida rápida'],
    'Mesero': ['Juan', 'Ana', 'Pedro', 'Ana', np.nan, 'Juan', 'Pedro', 'Ana'],
    'Fecha': ['2022-07-01', '2022-07-01', '2022-07-02', np.nan, '2022-07-02', '2022-07-03', '2022-07-03', '2022-07-04']
}

df = pd.DataFrame(data)

# Mostrar el DataFrame, recuerda al no usar print, se muestra de forma más amigable, esto es aplicable para todos los ejemplos.
df.head(10)
```

```python
nan_por_columna = df.isnull().sum()
nan_por_columna

nan_por_fila = df.isnull().sum(axis=1)
nan_por_fila
```

```python
df_limpieza_nans = df.copy()
df_limpieza_nans_columnas = df.copy()
df_limpieza_nans_subset = df.copy()
df_limpieza_fillna = df.copy()
df
```

```python
# any = si hay un NaN en la fila, se elimina
# all = si todos los valores son NaN en la fila, se elimina
df_nuevo = df_limpieza_nans.dropna(axis=0, how='all')
df_nuevo
```

```python
# any = si hay un NaN en la columna, se elimina
# all = si todos los valores son NaN en la columna, se elimina
df_nuevo_2 = df_limpieza_nans_columnas.dropna(axis=1, how='all')
df_nuevo_2
```

```python
df_nuevo_3 = df_limpieza_nans_subset.dropna(subset=['Categoría', 'Mesero'], how='any')
df_nuevo_3
```
