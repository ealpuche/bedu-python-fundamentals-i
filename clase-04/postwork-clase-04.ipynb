```python
"""
Imagina la siguiente situación: Eres el encargado de organizar un evento de capacitación para tu empresa y necesitas llevar un registro de asistencia para garantizar una adecuada planificación y seguimiento del evento.

Tu tarea es desarrollar un programa en Python que permita a los participantes registrar su asistencia al evento. El programa deberá permitir ingresar el nombre y apellido de los participantes, así como registrar automáticamente la fecha y hora de llegada. Una vez completado el registro, el programa deberá mostrar un mensaje de confirmación y permanecer en ejecución hasta que se decida finalizar.

Acciones que el programa debe realizar:

Permitir al usuario registrar su asistencia ingresando su nombre y apellido.
Registrar automáticamente la fecha y hora de llegada del participante.
Mostrar un mensaje de confirmación una vez completado el registro.
Mantener el programa en ejecución continua hasta que el usuario decida salir.
"""

import datetime

# print(f"El dia de hoy es: {datetime.datetime.now()}")

opcion_1 = "1. Registrar asistencia"
opcion_2 = "2. Mostrar asistentes"
opcion_3 = "3. Salir"

opcion_usuario = 0
asistentes = []

while opcion_usuario != 3:
    try:
        opcion_usuario = int(input(f"""
            Bienvenido al sistema de registro de asistencia.
            Por favor, seleccione una opcion:
            {opcion_1}
            {opcion_2}
            {opcion_3}
        """))
    except ValueError:
        print("Opcion invalida, intente de nuevo.")
        continue
    except:
        print("Ha ocurrido un error inesperado.")
        continue
    finally:
        print("===============================================")

    if opcion_usuario == 1:
        ## La tupla va tener el nombre, apellido y la fecha y hora de llegada
        nombre = input("Ingrese su nombre: ")
        apellido = input("Ingrese su apellido: ")
        asistentes.append((nombre, apellido, datetime.datetime.now()))
        print("Asistencia registrada con exito!")
        print("===============================================")

    elif opcion_usuario == 2:
        for asistente in asistentes:
            print(f"Nombre: {asistente[0]} {asistente[1]} - Fecha y hora de llegada: {asistente[2]}")
        print("===============================================")

    elif opcion_usuario == 3:
        print("Saliendo del programa...")
        break

    else:
        print("Opcion invalida, intente de nuevo.")
```
