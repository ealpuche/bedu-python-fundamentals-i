```python
"""
El programa debe realizar las siguientes acciones:

Solicitará al usuario seleccione que una opción: Registrar nuevo proyecto, actualizar estado de proyecto, mostrar lista de proyectos o salir.
Dependiendo la opción, solicitará al usuario el nombre del proyecto.
Pedirá al usuario que ingrese el estado del proyecto (no iniciado, en progreso, completado).
Permitirá al usuario actualizar el estado de los proyectos registrados.
Mostrará una lista de todos los proyectos con su estado actual.
Finalmente, el programa debe mostrar un mensaje con el nombre y el estado de todos los proyectos registrados cuando el usuario decida mostrar la lista de proyectos.
"""

opcion_1 = "Registrar nuevo proyecto"
opcion_2 = "Actualizar estado de proyecto"
opcion_3 = "Mostrar lista de proyectos"
opcion_4 = "Salir"

estado_1 = "no iniciado"
estado_2 = "en progreso"
estado_3 = "completado"

estados = {
    "1": "no iniciado",
    "2": "en progreso",
    "3": "completado"
}

proyectos = {}

opcion_usuario = 0
## desarrollo chedraui
## desarrollo soriana
## desarrollo walmart
while True:
    opcion_usuario = input(f"Seleccione una opción:\n1. {opcion_1}\n2. {opcion_2}\n3. {opcion_3}\n4. {opcion_4}\n")
    if opcion_usuario == "1":
        nombre_proyecto = input("Ingrese el nombre del proyecto: ")
        estado_proyecto = input(f"Ingrese el estado del proyecto: 1){estado_1}, 2){estado_2}, 3){estado_3}")
        proyectos[nombre_proyecto] = estados[estado_proyecto]
    elif opcion_usuario == "2":
        nombre_proyecto = input("Ingrese el nombre del proyecto a actualizar:")
         ##Validar que el proyecto existe
        proyectos_existentes = proyectos.keys()
        if nombre_proyecto not in proyectos_existentes:
            print("El proyecto no existe")
            continue
        estado_proyecto = input(f"Ingrese el nuevo estado del proyecto: 1){estado_1}, 2){estado_2}, 3){estado_3}")
        proyectos[nombre_proyecto] = estados[estado_proyecto]
    elif opcion_usuario == "3":
        print("==============================================")
        for proyecto, estado in proyectos.items():
            print(f"Proyecto: {proyecto}, Estado: {estado}")
        
        print("==============================================")
    elif opcion_usuario == "4":
        print("Adiós!")
        break
    else:
        print("Opción no válida")
```
