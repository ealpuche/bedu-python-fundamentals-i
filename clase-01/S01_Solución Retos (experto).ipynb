ğŸ† Nos vemos en el siguiente reto, Â¡mucho Ã©xito!.

````python
# Programa para simular la compra en una tienda de conveniencia sobre artÃ­culos.

# Solicitar al cliente el precio de 5 productos.

# Producto 1
precio1 = 20.5
stock1 = 10

# Producto 2
precio2 = 35.0
stock2 = 5

# Producto 3
precio3 = 10.0
stock3 = 20

# Producto 4
precio4 = 50.0
stock4 = 2

# Producto 5
precio5 = 25.0
stock5 = 15

# Comparando precios: Â¿Es el precio del producto 4 mayor que el precio del producto 2?
es_mas_caro = precio4 > precio2
print(f"\nÂ¿El precio del producto 4 es mayor que el del producto 2?: {es_mas_caro}")

# Usando operadores lÃ³gicos: Â¿Tienen el producto 3 y el producto 5 precios menores a 30 y stocks mayores a 10?
condicion_combinada = (precio3 < 30 and stock3 > 10) and (precio5 < 30 and stock5 > 10)
print(f"\nÂ¿Tienen el producto 3 y el producto 5 precios menores a 30 y stocks mayores a 10?: {condicion_combinada}")

# Operador relacional: Â¿Es el stock del producto 1 igual al del producto 5?
mismo_stock = stock1 == stock5
print(f"\nÂ¿El stock del producto 1 es igual al del producto 5?: {mismo_stock}")

# Operador lÃ³gico combinado con relacional: Â¿Es el stock del producto 2 menor que el del producto 1 y el precio del producto 2 mayor que 20?
condicion_compleja = stock2 < stock1 and precio2 > 20
print(f"\nÂ¿Es el stock del producto 2 menor que el del producto 1 y el precio del producto 2 mayor que 20?: {condicion_compleja}")

# Usando operadores de asignaciÃ³n para aumentar el precio del producto 1 en un 10%
precio1 += precio1 * 0.10  # Aumentar el precio en un 10%
print(f"\nNuevo precio del producto 1 despuÃ©s del aumento: ${precio1:,.2f}")

# Total a pagar
total = precio1 + precio2 + precio3 + precio4 + precio5
print(f"\nTotal a pagar: ${total:,.2f}")
````

````python
# Resolver una actividad, simulando una cotizaciÃ³n para la compra de un automÃ³vil, con el fin de obtener una pequeÃ±a tabla de amortizaciÃ³n sobre los datos.
# Programa para la cotizaciÃ³n y financiamiento de un automÃ³vil.

# Solicitar el modelo del automÃ³vil para una mejor interacciÃ³n
modelo_automovil = input("Introduce el modelo del automÃ³vil que deseas comprar: ")

# Ingresa los datos del automÃ³vil y el financiamiento que deseas.
precio = float(input("Introduce el precio del automÃ³vil que deseas comprar: $"))


# Solicitar el nÃºmero de meses para el financiamiento
meses = int(input("Â¿A cuÃ¡ntos meses deseas el financiamiento del automÃ³vil? (Ejemplo: 12, 24, 36): "))

# Calcular la cuota mensual utilizando un operador de asignaciÃ³n
cuota_mensual = 0
cuota_mensual += precio / meses  # Equivalente a cuota_mensual = cuota_mensual + precio / meses

print("-"*60)

# Mostrar detalles de financiamiento utilizando interpolaciÃ³n de strings
print(f"\nHas seleccionado el modelo '{modelo_automovil}'.")
print(f"El precio total es: $ {precio:,.2f}")
print(f"Financiando a {meses} meses, tu cuota mensual serÃ¡ de: $ {cuota_mensual:,.2f}")

print("-"*60)

# Mostrar una tabla de amortizaciÃ³n simple para los primeros dos meses
print("\nTabla de AmortizaciÃ³n (primeros dos meses):")
print(f"Mes 1: Pago de $ {cuota_mensual:,.2f}, saldo pendiente: $ {precio - cuota_mensual:,.2f}")
print(f"Mes 2: Pago de $ {cuota_mensual:,.2f}, saldo pendiente: $ {precio - 2 * cuota_mensual:,.2f}")

print("-"*60)

# Mostrar un mensaje de despedida
print("\nÂ¡Gracias por utilizar nuestro sistema de financiamiento!")
````
